```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from plotnine import *
from plotnine.data import *

nasadf = pd.read_csv('https://data.giss.nasa.gov/tmp/gistemp/STATIONS/tmp_USW00014733_14_0_1/station.csv')
nasadf.replace(999.90, np.NaN,inplace=True) # 替换数值
nasadf.head()
```

```python
# Gallery Plot
(
    ggplot()
    + geom_line(nasadf, aes(x='YEAR', y='D-J-F'),colour = 'r')
    + geom_line(nasadf, aes(x='YEAR', y='M-A-M'),colour = 'g')
    + geom_line(nasadf, aes(x='YEAR', y='J-J-A'),colour = 'b')
    + geom_line(nasadf, aes(x='YEAR', y='S-O-N'),colour = 'y')
    + geom_smooth(nasadf, aes(x='YEAR', y='D-J-F'),method='loess',colour = 'r') #平滑曲线基于loess规则
    + geom_smooth(nasadf, aes(x='YEAR', y='M-A-M'),method='loess',colour = 'g')
    + geom_smooth(nasadf, aes(x='YEAR', y='J-J-A'),method='loess',colour = 'b')
    + geom_smooth(nasadf, aes(x='YEAR', y='S-O-N'),method='loess',colour = 'y')
    + labs(x='Years', y='℃')
    + theme (figure_size = (16,30))
)
```

```python
plt.figure(figsize=(16,6)) # 画布宽和高
plt.title("Daily Global Streams of Popular Songs in 2017-2018") # 图的标题
plt.xlabel("Airline") # x轴标签
plt.ylabel("Arrival delay (in minutes)") # y轴标签
plt.legend() # 强制显示图例
sns.set_style("dark") # 画布背景设置，有darkgrid、whitegrid、dark、white、ticks5种
sns.lineplot(data=nasadf['D-J-F'], label="D-J-F")
sns.lineplot(data=nasadf['M-A-M'], label="M-A-M")
sns.lineplot(data=nasadf['J-J-A'], label="J-J-A")
sns.lineplot(data=nasadf['S-O-N'], label="S-O-N")
```
